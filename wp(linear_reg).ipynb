```python
#renaming our columns
small_df= small_df.rename(columns={'location_name':'location','last_updated':'date','condition_text':'cloud_cover'})
small_df.head() 
```

```python
# Converting 'date' column to datetime format
small_df['date_time'] = pd.to_datetime(small_df['date'], format='%Y-%m-%d %H:%M')
small_df.head()
```

```python
#renaming our columns
small_df= small_df.rename(columns={'location_name':'location','last_updated':'date','condition_text':'cloud_cover'})
small_df.head() 
```

```python
small_df.columns
```

```python
#ordinal_encoder on condition_text
col=['air_quality_Carbon_Monoxide', 'air_quality_Ozone',
       'air_quality_Nitrogen_dioxide', 'air_quality_Sulphur_dioxide',
       'air_quality_PM2.5', 'air_quality_PM10', 'air_quality_us-epa-index',
       'air_quality_gb-defra-index']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
transform=ColumnTransformer([('oe',OrdinalEncoder(),['cloud_cover']),
                             ('ss',StandardScaler(),col)],remainder='passthrough')
```

```python
transform.fit(small_df)
```

```python
new_col=['cloud_cover','air_quality_Carbon_Monoxide', 'air_quality_Ozone', 'air_quality_Nitrogen_dioxide', 'air_quality_Sulphur_dioxide', 'air_quality_PM2.5', 'air_quality_PM10', 'air_quality_us-epa-index', 'air_quality_gb-defra-index','location', 'region', 'latitude', 'longitude', 'date', 'temperature_celsius', 'wind_kph', 'pressure_in', 'precip_mm', 'humidity', 'cloud', 'visibility_km', 'date_time']
```

```python
data=transform.fit_transform(small_df)
```

```python
transform_data=pd.DataFrame(data,columns=new_col)
```

```python
transform_data.head(3)
```

```python
#visualization...
```

```python
# selecting the independent feature i.e., features in which the ML model will be trained 
independent_feature= ['wind_kph','pressure_in','precip_mm','humidity','air_quality_Carbon_Monoxide','air_quality_Ozone','air_quality_Nitrogen_dioxide','air_quality_Sulphur_dioxide','air_quality_PM2.5','air_quality_PM10','air_quality_us-epa-index','air_quality_gb-defra-index'] 
# selecting the targeted feature i.e., features that will be predicted based on their respective value in independent feature
targeted_feature= ['temperature_celsius']
```

```python
X=transform_data[independent_feature]
Y=transform_data[targeted_feature]
```

```python
# data spliting
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size= 0.2, random_state=42)
 
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_test.shape)
print(y_train.shape)
```

```python
from sklearn.model_selection import cross_val_score
#linear regression
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
cross_val_results = cross_val_score(lr,X_train,y_train)
```

```python
cross_val_results
```

```python
#mean accuracy of our model is
np.mean(cross_val_results)
```

```python
lr.fit(X_train,y_train)
```

```python
#predicting data by the help of linearRegression
y_pred=lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```
