```python
import pandas as pd
#'weather' is variable for our dataset
weather = pd.read_csv('historic1_weather_data.csv')
weather
```

```python
#renaming our columns
weather = weather.rename(columns={'Temperature (C)':'temperature'})
weather = weather.rename(columns={'Precipitation (mm)':'precipitation'})
weather = weather.rename(columns={'Wind Speed (km/h)':'wind_speed'})
weather = weather.rename(columns={'Humidity':'humidity'})
weather.head()
```

```python
# Convert 'date','time' column to datetime format
weather['date'] = pd.to_datetime(weather['Date'], dayfirst=True)
weather['time'] = pd.to_datetime(weather['Time'], format='%H:%M')
```

```python
#Now we will calculate the mean , varience and standard deviation for further applying
# with columns 'temperature', 'humidity','precipitation' and 'wind_speed'

# Calculate mean, variance, and standard deviation of temperature
temperature_mean = weather['temperature'].mean()
temperature_variance = weather['temperature'].var()
temperature_std = weather['temperature'].std()

# Calculate mean, variance, and standard deviation of humidity
humidity_mean = weather['humidity'].mean()
humidity_variance = weather['humidity'].var()
humidity_std = weather['humidity'].std()

# Calculate mean, variance, and standard deviation of wind speed
wind_speed_mean = weather['wind_speed'].mean()
wind_speed_variance = weather['wind_speed'].var()
wind_speed_std = weather['wind_speed'].std()

# Calculate mean, variance, and standard deviation of temperature
precipitation_mean = weather['precipitation'].mean()
precipitation_variance = weather['precipitation'].var()
precipitation_std = weather['precipitation'].std()

print("Temperature - Mean: {}, Variance: {}, Standard Deviation: {}".format(temperature_mean, temperature_variance, temperature_std))
print("Humidity - Mean: {}, Variance: {}, Standard Deviation: {}".format(humidity_mean, humidity_variance, humidity_std))
print("Wind Speed - Mean: {}, Variance: {}, Standard Deviation: {}".format(wind_speed_mean, wind_speed_variance, wind_speed_std))
print("Precipitation - Mean: {}, variance: {}, Standard Deviation: {}".format(precipitation_mean, precipitation_variance, precipitation_std))
```

```python
#Now we will calculate Temporal aggregation which we will use further 
# Calculate daily average temperature, humidity, and total precipitation
daily_aggregated_data = weather.resample('D', on='date').agg({
    'temperature': 'mean',
    'humidity': 'mean',
    'precipitation': 'sum'
})

print("Daily Aggregated Data:")
print(daily_aggregated_data.head())
```

```python
# Now we will perform 'Moving Averages' for short term fluctuation in the dataset
# Calculate 7-day moving average of temperature
weather['temperature_7day_ma'] = weather['temperature'].rolling(window=7).mean()
weather['humidity_7day_ma'] = weather['humidity'].rolling(window=7).mean()
weather['wind_speed_7day_ma'] = weather['wind_speed'].rolling(window=7).mean()
print("Weather Data with Moving Averages:")
print(weather[['date', 'temperature', 'temperature_7day_ma','humidity','humidity_7day_ma','wind_speed','wind_speed_7day_ma']])
#Remember these are new columns in our dataset
```
