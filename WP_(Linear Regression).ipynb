```python
import pandas as pd
import  numpy  as np
```

```python
weather01 = pd.read_csv('IndianWeatherRepository 1.csv')
weather01.head()
```

```python
columns_to_drop =['country','timezone','last_updated_epoch','sunrise','sunset','moonrise','moonset','moon_phase','moon_illumination','wind_mph','wind_degree','wind_direction','pressure_mb','precip_in','feels_like_celsius','feels_like_fahrenheit','visibility_miles','uv_index','gust_mph','gust_kph','temperature_fahrenheit']
weather01 = weather01.drop(columns_to_drop, axis=1)
weather01.head()
```

```python
titles = list(weather01.columns)
titles
```

```python
titles[4],titles[19] = titles[19],titles[4]  
titles
```

```python
weather01 = weather01[titles]
weather01
```

```python
first_row = weather01.iloc[0]
print(first_row)
```

```python
#renaming our columns
weather01 = weather01.rename(columns={'location_name':'location','last_updated':'date','condition_text':'cloud_cover'})
weather01.head() 
```

```python
# Converting 'date' column to datetime format
weather01['date_time'] = pd.to_datetime(weather01['date'], format='%Y-%m-%d %H:%M')
weather01.head()
```

```python
#Now we will calculate Temporal aggregation which we will use further 
# Calculate daily average temperature, humidity, and total precipitation
daily_aggregated_data = weather01.resample('D', on='date_time').agg({
    'temperature_celsius': 'mean',
    'humidity': 'mean',
    'precip_mm': 'sum'
})

print("Daily Aggregated Data:")
print(daily_aggregated_data.head())
```

```python
#Grouping differet locations for further actions
grouped_data = weather01.groupby(['location'])
#you can repeat this function with different locations
Ashoknagar = grouped_data.get_group('Ashoknagar')
Ashoknagar
```

```python
Ashoknagar.index
Ashoknagar.reindex(columns=['location',	'region','date_time',	'latitude',	'longitude',	'air_quality_us-epa-index',	'cloud_cover',	'wind_kph',	'pressure_in',	'precip_mm','visibility_km',	'air_quality_Carbon_Monoxide',	'air_quality_Ozone',	'air_quality_Nitrogen_dioxide',	'air_quality_Sulphur_dioxide',	'air_quality_PM2.5','air_quality_PM10',	'date',	'air_quality_gb-defra-index','temperature_celsius'])
```
